```python
import numpy as np
import jax.numpy as jnp
import matplotlib.pyplot as plt
from minimum_spanning_tree_inverse_spline_render import get_splines_from_frame
from splender.image import fit
```

```python
dat = np.loadtxt('mnist_train.csv', delimiter=',')
```

```python
i = 2
img = dat[i, 1:].reshape(28, 28)
img = img / 255.0
path = get_splines_from_frame(img, 0.1)[0][:, ::-1]
print(len(path))
init_spline = np.array(path)
init_spline = init_spline[::(init_spline.shape[0] // 8)]
print(init_spline.shape)
plt.imshow(img, cmap='gray')
plt.plot(path[:, 0], path[:, 1], 'r-')
plt.plot(init_spline[:, 0], init_spline[:, 1], 'g-')
```

```python
(jnp.array([init_spline[None]] * 3)).shape
```

```python
n_tries = 3

splines, recon, _, _, _, _ = fit(jnp.array([img] * n_tries), jnp.array([init_spline[None]] * n_tries), s = 10)

print(len(splines))
print(splines[0].shape)
print(recon.shape)

# plot
for i in range(1, n_tries + 1):
    plt.subplot(n_tries, 2, 2 * i - 1)
    plt.imshow(img, cmap='gray', vmin=0, vmax=1)
    for spline in splines[i-1]:
        plt.plot(spline[:, 0], spline[:, 1], 'r-')
    plt.subplot(n_tries, 2, 2 * i)
    plt.imshow(recon[i-1], cmap='gray', vmin=0, vmax=1)
plt.show()
```
